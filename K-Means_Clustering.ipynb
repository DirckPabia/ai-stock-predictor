```python
#K-Means Clustering
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 

# Allows charts to appear in the notebook 
%matplotlib inline
```

```python
# Import dataset
dataset = pd.read_csv(r'C:\Users\Dirck\OneDrive\Documents\Machine Learning Tutorial Part 1 _ 2\Part 2\cars (dataset for k means).csv')
```

```python
X = dataset[:-1]
X.head()
```

```python
# Eliminating null values
for i in X.select_dtypes(include=['float64', 'int64']).columns:
    X.loc[:, i] = X[i].fillna(X[i].mean())

#for i in X.select_dtypes(include=['object']).columns:
#    X.loc[:, i] = X[i].fillna(X[i].mode()[0])


for i in X.columns:
    print(X[i].isnull().sum())
```

```python
# Strip whitespace and replace blank strings with NaN
X = X.replace(r'^\s*$', np.nan, regex=True)

# Now handle missing values again
for col in X.select_dtypes(include=['float64', 'int64']).columns:
    X.loc[:, col] = X[col].fillna(X[col].mean())

for col in X.select_dtypes(include=['object']).columns:
    X.loc[:, col] = X[col].fillna(X[col].mode()[0])

# Convert all remaining non-numeric columns to numbers if needed (e.g. label encoding)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in X.select_dtypes(include=['object']).columns:
    X.loc[:, col] = le.fit_transform(X[col])
```

```python
#Using the elbow method to find the optimal number of clusters

from sklearn.cluster import KMeans

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Applying k-means to the cars dataset 
kmeans = KMeans(n_clusters=3,init='k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans = kmeans.fit_predict(X)
X = X.to_numpy()
```

```python
plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=100,c='red',label='Toyota')
plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=100,c='blue',label='Nissan')
plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=100,c='green',label='Honda')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='yellow',label='Centroids')
plt.title('CLuster of car make')
plt.legend()
plt.show()
```
